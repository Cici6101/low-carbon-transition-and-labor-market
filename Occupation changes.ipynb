```python
import matplotlib.pyplot as plt
import pandas as pd

# 读取 Excel 数据
df = pd.read_excel("OCCUPATION.xlsx", sheet_name="total")
df.set_index("Occupation", inplace=True)
df = df.T  # 年份为行，职业为列

# 定义 9 种风格（线型、颜色、标记点）
line_styles = ['-', '-', '-','-',':','-', ':', '--',  '-.']
markers = ['o', 's', '^', 'D', 'o', None, None, None, None]

# 创建图形
plt.figure(figsize=(10, 7))

# 绘制每个职业的线
for i, column in enumerate(df.columns):
    plt.plot(
        df.index, df[column],
        linestyle=line_styles[i],
        marker=markers[i],
        color="black",
        label=column,
        linewidth=1.5,
        markersize=4
    )

# 添加标题和标签
plt.axhline(y=0, color='gray', linewidth=1, linestyle='--')

plt.ylabel('Occupational Employment Change (thousands)',fontsize=14)
plt.legend()
plt.tight_layout()
plt.savefig('Occupation Change1.png', dpi=300)
# 显示图形
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# === 1. 读取数据 ===
df = pd.read_excel("OCCUPATION.xlsx", sheet_name="Global")

# === 2. 添加排序列，确保保留 Excel 原顺序 ===
df['row_order'] = range(len(df))
df = df.sort_values(by=['sector', 'row_order'])

# === 3. 数值列（年份） ===
value_columns = [col for col in df.columns if col not in ['sector', 'Occupation', 'row_order']]

# === 4. 样式定义 ===
line_styles = ['-', '-', '-', '-', ':', '-', ':', '--', '-.']
markers = ['o', 's', '^', 'D', 'o', None, None, None, None]

highlight_sectors = ['B', 'C', 'D', 'F']
highlight_colors = {
    'B': 'orchid',
    'C': 'lightskyblue',
    'D': 'lightgreen',
    'F': 'lightcoral'
}
default_color = 'lightgray'

sector_labels = {
    'A': 'Agriculture, forestry and fishing',
    'B': 'Mining and quarrying',
    'C': 'Manufacturing',
    'D': 'Electricity, gas, steam and air conditioning supply',
    'E': 'Water supply; sewerage, waste management and remediation',
    'F': 'Construction',
    'G': 'Wholesale and retail trade; repair of motor vehicles and motorcycles',
    'H': 'Transportation and storage',
    'I': 'Accommodation and food service activities',
    'J': 'Information and communication',
    'K': 'Financial and insurance activities',
    'L': 'Real estate activities',
    'M': 'Professional, scientific and technical activities',
    'N': 'Administrative and support service activities',
    'O': 'Public administration and defence; compulsory social security',
    'P': 'Education',
    'Q': 'Human health and social work activities',
    'R': 'Arts, entertainment and recreation',
    'S': 'Other service activities',
    'T': 'Activities of households',
    'U': 'Activities of extraterritorial organisations and bodies'
}

# === 5. 开始绘图 ===
plt.figure(figsize=(10, 7))
added_labels = set()

for i, row in df.iterrows():
    sector = row['sector']
    occupation = row['Occupation']
    color = highlight_colors.get(sector, default_color)

    # 图例标签设置（只加一次）
    if sector in highlight_sectors:
        sector_name = sector_labels.get(sector, sector)
        label = sector_name if sector_name not in added_labels else None
    else:
        label = "Other sectors" if "Other sectors" not in added_labels else None

    if label:
        added_labels.add(label)

    # 绘图数据
    x = value_columns
    y = row[value_columns].astype(float).values

    plt.plot(
        x, y,
        label=label,
        linestyle=line_styles[i % len(line_styles)],
        marker=markers[i % len(markers)],
        color=color,
        linewidth=1.5,
        markersize=4 if markers[i % len(markers)] else 0,
        alpha=1.0 if sector in highlight_sectors else 0.3
    )

# === 6. 美化输出 ===
plt.axhline(y=0, color='gray', linewidth=1, linestyle='--')
plt.ylabel('Detailed Occupational Employment Change(thousands)', fontsize=14)
plt.legend()
plt.tight_layout()
plt.savefig('Occupation_Change_Global.png', dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# Read data
df = pd.read_excel("SKILL.xlsx")
df_positive= df.dropna(subset=['weighted_skill_positive'])
df_negative= df.dropna(subset=['weighted_skill_negative'])


# List of years to plot
years = [2030, 2040, 2050]

# Create subplots (1 row, 3 columns)
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharex=True, sharey=True)

# Loop through each year and plot on the corresponding axis
for i, year in enumerate(years):
    # Select data for specific year
    pos_df = df_positive[df_positive['year'] == year]
    neg_df = df_negative[df_negative['year'] == year]
    positive_skills = pos_df['weighted_skill_positive']
    negative_skills = neg_df['weighted_skill_negative']

    # Perform a t-test (Independent two-sample t-test)
    t_stat, p_value = stats.ttest_ind(positive_skills, negative_skills)

    print(f"T-statistic: {t_stat}")
    print(f"P-value: {p_value}")


    # Prepare positive values part
    pos_df = pos_df[['weighted_skill_positive', 'total_change_positive']].copy()
    pos_df.columns = ['skill', 'jobs']
    pos_df['type'] = 'Increase'
    pos_df['jobs'] = pos_df['jobs'] / 1000 

    # Prepare negative values part
    neg_df = neg_df[['weighted_skill_negative', 'total_change_negative']].copy()
    neg_df.columns = ['skill', 'jobs']
    neg_df['type'] = 'Decrease'
    neg_df['jobs'] = neg_df['jobs'] / 1000 

    # Combine both dataframes
    plot_df = pd.concat([pos_df, neg_df])

    # Plotting KDE for Increase category and Decrease category
    sns.kdeplot(data=pos_df, x='skill', label="Increase", fill=True, alpha=0.6, common_norm=False, color='green', ax=axes[i])
    sns.kdeplot(data=neg_df, x='skill', label="Decrease", fill=True, alpha=0.6, common_norm=False, color='orange', ax=axes[i])
    ax_right = axes[i].twinx()  # Create a second y-axis
    sns.histplot(data=plot_df, x='skill', hue='type', bins=20, multiple='stack', kde=False, ax=ax_right, palette='Set2', alpha=0.6)

    # Title and labels for each subplot
    axes[i].set_xlabel("Skill Level")
    axes[i].set_ylabel("Probability Density Function")
    axes[i].grid(True, linestyle='--', alpha=0.3)

    # Right Y-axis for job counts

    axes[i].set_xlim(0.5, 4)
    ax_right.set_ylim(0, 40)
    if i == 2:
        ax_right.set_ylabel("Number of Countries", color='black', rotation=270, labelpad=15)
    else:
        ax_right.set_ylabel('')  # Remove the right Y-axis label for the first two subplots

# Adjust layout
plt.tight_layout()
plt.savefig("Probability Density Function.png")
plt.show()
```
